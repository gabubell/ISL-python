## Question 9 - Auto Data
### (a)

```python
auto = islp.load_data('Auto')
auto.head()
```

```python
datatypes = {'quant': ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration'],
             'qual': ['origin', 'name']}

auto_quant = auto[datatypes['quant']].astype(float)
```

```python
auto_quant.max()
```

### (b) and (c)

```python
def range(df):
    df_out = pd.DataFrame({
        'range': df.max() - df.min(),
        'median': df.median(),
        'mean': df.mean(),
        'std': df.std(),
        }
    ).round(2)

    return df_out

range(auto_quant)
```

### (d)

```python
auto_quant_filt = auto_quant.drop(labels = auto_quant.index[10:85])
auto_quant_filt
```

```python
range(auto_quant_filt)
```

### (e)

```python
#remove cylinders from plot because it is discrete with few observations
#we can consider it to be a categorical (qualitative) variable
pd.plotting.scatter_matrix(auto_quant.loc[:, auto_quant.columns != 'cylinders']);
```

```python
#Applying natural log to all cols
#we can notice the relationship between the variables approximate more to a linear one
auto_quant_ln = auto_quant.assign(
    **{col: np.log(auto_quant[col]) for col in auto_quant.columns}
)

pd.plotting.scatter_matrix(auto_quant_ln.loc[:, auto_quant_ln.columns != 'cylinders']);
```

### (f)

```python
corr = auto_quant_ln.corr()
corr.style.background_gradient(cmap='coolwarm')
```

We can notice both from scatter and correlation matriz that mpg has a linear correlation with all variable, and only acceleration being positive. So, we can  test the usage of all variables to predict mpg with a linear model.

## Question 10 - Boston Housing Data
### (a) and (b)

```python
housing = islp.load_data('Boston')
housing.head()
```

```python
housing.describe()
```

The Boston housing dataset has 506 rows, each one of them representing a suburb of Boston. According to [ISLP reference](https://intro-stat-learning.github.io/ISLP/datasets/Boston.html), we have the following meaning for each column: 
- crim: per capita crime rate by town.
- zn: proportion of residential land zoned for lots over 25,000 sq.ft.
- indus: proportion of non-retail business acres per town.
- chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
- nox: nitrogen oxides concentration (parts per 10 million).
- rm: average number of rooms per dwelling.
- age: proportion of owner-occupied units built prior to 1940.
- dis: weighted mean of distances to five Boston employment centres.
- rad: index of accessibility to radial highways.
- tax: full-value property-tax rate per $10,000.
- ptratio: pupil-teacher ratio by town.
- lstat: lower status of the population (percent).
- medv: median value of owner-occupied homes in $1000s.

### (c) and (d)

```python
subset_cols = ['crim', 'tax', 'ptratio', 'rm', 'age', 'dis', 'rad', 'medv']
pd.plotting.scatter_matrix(housing.loc[:, subset_cols]);
```

```python
housing_ln = housing.assign(
    **{col: np.log(housing[col]) for col in housing.columns}
);

pd.plotting.scatter_matrix(housing_ln.loc[:, subset_cols]);
```

```python
#Note that the warning when passing ln is because min(crim) is very small
[col for col in subset_cols if (housing[col] < 0.01).any()], housing['crim'].min()
```

```python
corr = housing_ln.loc[:, subset_cols].corr()
corr.style.background_gradient(cmap='coolwarm')
```

We can notice some degree of relationship between crim and all selected cols, which is stronger with rad (accesibility to radial highways), meaning that closer to radial highways more crimes are in the suburbs. Those suburbs also show lower levels of medv (media value of homes), noted by the negative linear correlation.

While on variables like crim, medv and ptration have low variance they show some suburbs that are outliers, meaning they distance a lot themselves from the others.
In the opposite, tax has high variance, where the top 50% of suburbs show a range from 279 to 711. While this variable doesn't show any outliers by inspecting the boxplot, it has a high range of values concentrated on the top 50%.
