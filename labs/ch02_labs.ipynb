```python
total = 0
for vl in [3, 2, 19]:
    total += vl
print('Total = {0}'.format(total))
```

```python
total = 0
for vl in [2, 3, 19]:
    for weight in [3, 2, 1]:
        total += vl * weight
        print(vl, weight, vl*weight)
print('Total is: {0}'.format(total))
```

```python
total = 0
for vl, weight in zip([2, 3, 19], [0.2, 0.3, 0.5]):
    total += vl * weight
    print(vl, weight, vl*weight, total)
print('Weighted average is: {0}'.format(total))
```

```python
#string formatting
rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5))
M = rng.choice([0, np.nan], p=[0.8, 0.2], size = A.shape)
A += M
D = pd.DataFrame(A, columns = ['food', 'bar', 'pickle', 'snack', 'popcorn'])
D[:3]
```
